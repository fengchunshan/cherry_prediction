Temperatures, rain and wind have a strong effect on the bloom date, for example, they can accelerate, delay or shorten it considerably.
Because we have difficulties to get future temperatures, rain and wind date, so we decide to use time series to forecast.

A time series is a series of data points in time order. Most commonly, a time series is a sequence taken at successive equally spaced points in time. Time series data are simply measurements or events that are tracked, monitored, down sampled, and aggregated over time. 

The biggest advantage for time series is that it only needs the historical data, and do not need other variables, so we can continue to use time series algorithm to forecast in future.

###  Use ARIMA time series model to predict for kyoto

```python
# evaluate an ARIMA model using a walk-forward validation
from pandas import read_csv
## from pandas import datetime
import pandas as pd
from matplotlib import pyplot
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
# load dataset
data = pd.read_csv('kyoto.csv')
data=data['bloom_doy']

# split into train and test sets
X = data.values
import warnings
warnings.filterwarnings('ignore')
history = [x for x in X]
predictions = list()
for t in range(10):
    model = ARIMA(history, order=(20,1,20))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
   
    print('predicted=%d' % (yhat))
    history.append(yhat)

pyplot.plot(predictions, color='red')
pyplot.show()
```
